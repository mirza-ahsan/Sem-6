```python
import random
```

```python
def initialisePopulation(n):
    population = []
    for i in range(1000):
        row = []
        for j in range(n):
            row.append(random.randint(1, n))
        population.append(row)

    return population
```

```python
def inAttack(i, solution):
    for j in range(len(solution)):
        if i == j:
            continue

        start_row = j
        start_col = solution[j] - 1

        end_row = i
        end_col = solution[i] - 1

        if start_col == end_col:
            return True

        if abs(start_row - end_row) == abs(start_col - end_col):
            return True

    return False
```

```python
def fitnessFunction(solution):
    count = 0
    for i in range(len(solution)):
        if inAttack(i, solution):
            count += 1

    return count
```

```python
def generatePopulation(parent_solutions):
    child_solutions = []
    n = len(parent_solutions[0][1])

    for i in range(0, len(parent_solutions), 2):
        p1 = parent_solutions[i][1]
        p2 = parent_solutions[i + 1][1]

        crossover_point = random.randint(1, n - 1)
        c1 = p1[:crossover_point] + p2[crossover_point:]
        c2 = p2[:crossover_point] + p1[crossover_point:]

        if random.random() < 0.8:
            c1[random.randint(0, n - 1)] = random.randint(1, n)
        if random.random() < 0.8:
            c2[random.randint(0, n - 1)] = random.randint(1, n)

        child_solutions.append((fitnessFunction(c1), c1))
        child_solutions.append((fitnessFunction(c2), c2))

    child_solutions.sort()
    parent_solutions.sort()

    half_len = len(parent_solutions) // 2
    next_population = parent_solutions[:half_len] + child_solutions[:half_len]

    next_population.sort()

    return next_population


def main():
    n = 8
    population = initialisePopulation(n)

    solutions = []
    for sol in population:
        solutions.append((fitnessFunction(sol), sol))

    solutions.sort()

    for generation in range(1000):
        if solutions[0][0] == 0:
            print(f"Perfect solution found in Generation {generation}")
            break

        solutions = generatePopulation(solutions)

    print(f"Best fitness score (0 is perfect): {solutions[0][0]}")
    print(f"Winning Board Array: {solutions[0][1]}")


if __name__ == "__main__":
    main()
```
