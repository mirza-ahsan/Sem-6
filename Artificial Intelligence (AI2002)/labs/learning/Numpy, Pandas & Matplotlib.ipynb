## Array Operations

```python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)
```

```python
print(x + y)
print(np.add(x, y))
print(x - y)
print(np.subtract(x, y))
print(x * y)
print(np.multiply(x, y))
print(x / y)
print(np.divide(x, y))
print(np.sqrt(x))
```

## Display Multiple Plots

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```

plt.subplot(nrows, ncols, index)

nrows: number of rows of plots.

ncols: no of columns of plots.

index: which plot you are working on.

```python
x = np.linspace(0, 10, 100)
for i in range(1, 5):  # 4 plots
    plt.subplot(2, 2, i)  # 2 rows, 2 columns
    plt.plot(x, np.sin(x + i))  # shifting sine wave
    plt.title(f"Plot {i}")

plt.tight_layout()
plt.show()
```

When you create multiple subplots, Matplotlib often overlaps plot elements (like titles, labels, and axes) because it doesn't automatically adjust for spacing. plt.tight_layout() automatically adjusts the spacing between subplots so that the titles don’t overlap with other plots, X/Y axis labels don’t get cut off and the layout looks clean and readable

## Customizing Plots

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.figure(figsize=(8, 5))#width,height
plt.plot(x, y, color='green', linestyle='--', marker='o', label="Line 1")
plt.title("Customized Line Plot", fontsize=14)
plt.xlabel("X-Axis", fontsize=12)
plt.ylabel("Y-Axis", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4]

# First line data
y1 = [10, 20, 25, 30]
# Second line data
y2 = [15, 18, 22, 28]
plt.figure(figsize=(8, 5))

plt.plot(x, y1, color='green', linestyle='--', marker='o', label="Line 1")
plt.plot(x, y2, color='blue', linestyle='-', marker='s', label="Line 2")

plt.title("Customized Multi-Line Plot", fontsize=14)
plt.xlabel("X-Axis", fontsize=12)
plt.ylabel("Y-Axis", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.show()
```

```python
labels = ['Python', 'Java', 'C++']
sizes = [40, 30, 30]
colors = ['skyblue', 'darkblue', 'lightblue']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
plt.legend()
plt.show()
```

The startangle parameter in plt.pie() sets the starting angle (in degrees) where the first wedge (i.e. the first slice) is drawn measured counter-clockwise from the x-axis (which points to the right).

## Plotting from a Data Frame

```python
import pandas as pd
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [250, 300, 280, 350, 400],
    'Expenses': [200, 220, 210, 300, 330]
}

df = pd.DataFrame(data)
df
```

Plot Sales and Expenses over Months

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='s', label='Expenses')
plt.title('Monthly Sales and Expenses', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

## Correlation Matrix

```python
import seaborn as sns  # for a nice heatmap
```

```python
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100) * 50 + 20,
    'C': np.random.randn(100) * 10 - 10,
    'D': np.random.randn(100) + 5
})

data
```

```python
corr = data.corr()
print(corr)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```

```python
plt.matshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.matshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}",
                 ha='center', va='center', color='black')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
