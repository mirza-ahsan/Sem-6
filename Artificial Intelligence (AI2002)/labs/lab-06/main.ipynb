```python
import heapq as hq
```

```python
def readCube(filename):
    cube = []
    with open(filename, "r") as file:
        for line in file:
            cube.append(line.strip().split())
    return cube
```

```python
def getPosition(cube, target):
    for r in range(len(cube)):
        for c in range(len(cube[r])):
            if cube[r][c] == target:
                return (r, c)
    return None
```

```python
def getManhattenHeuristic(current, goal):
    curr_r, curr_c = current
    goal_r, goal_c = goal
    # h(n) = |x1 - x2| + |y1 - y2|
    return abs(goal_r - curr_r) + abs(goal_c - curr_c)
```

# A* Search

```python
cube = readCube("cube.txt")
start_pos = getPosition(cube, "S")
goal_pos = getPosition(cube, "G")

pq = []
tie_breaker = 0
path = [start_pos]
path_cost = 0

hq.heappush(
    pq,
    (
        path_cost + getManhattenHeuristic(start_pos, goal_pos),
        tie_breaker,
        start_pos,
        path,
        path_cost,
    ),
)

visited_g_costs = {start_pos: 0}

final_path = None

while len(pq) != 0:
    f, _, node_pos, node_path, gn = hq.heappop(pq)
    r1, c1 = node_pos

    if node_pos == goal_pos:
        final_path = node_path
        break

    neighbors = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for r2, c2 in neighbors:
        nr = r1 + r2
        nc = c1 + c2

        if not (0 <= nr < len(cube) and 0 <= nc < len(cube[0])):
            continue

        cell_value = cube[nr][nc]

        if cell_value == "1":
            continue

        step_cost = 1
        if cell_value == "2" or cell_value == "+":
            step_cost = 10

        new_gn = gn + step_cost

        if (nr, nc) not in visited_g_costs or new_gn < visited_g_costs[(nr, nc)]:
            visited_g_costs[(nr, nc)] = new_gn

            new_path = list(node_path)
            new_path.append((nr, nc))

            hn = getManhattenHeuristic((nr, nc), goal_pos)
            fn = new_gn + hn

            tie_breaker += 1
            hq.heappush(pq, (fn, tie_breaker, (nr, nc), new_path, new_gn))

if final_path:
    print(f"Path found with {len(final_path)-1} steps")
    print("Coordinates:", final_path)
else:
    print("No valid path exists.")
```

# Greedy Best First Search

```python
cube = readCube("cube.txt")
start_pos = getPosition(cube, "S")
goal_pos = getPosition(cube, "G")

pq = []
tie_breaker = 0
path = [start_pos]
path_cost = 0

initial_h = getManhattenHeuristic(start_pos, goal_pos)
hq.heappush(
    pq,
    (
        initial_h,
        tie_breaker,
        start_pos,
        path,
        path_cost,
    ),
)

visited = set([start_pos])

final_path = None
final_cost = 0

while len(pq) != 0:
    hn, _, node_pos, node_path, gn = hq.heappop(pq)
    r1, c1 = node_pos

    if node_pos == goal_pos:
        final_path = node_path
        final_cost = gn
        break

    neighbors = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for r2, c2 in neighbors:
        nr = r1 + r2
        nc = c1 + c2

        if not (0 <= nr < len(cube) and 0 <= nc < len(cube[0])):
            continue

        cell_value = cube[nr][nc]

        if cell_value == "1":
            continue

        if (nr, nc) in visited:
            continue

        step_cost = 1
        if cell_value == "2" or cell_value == "+":
            step_cost = 10

        new_gn = gn + step_cost

        visited.add((nr, nc))

        new_path = list(node_path)
        new_path.append((nr, nc))

        new_hn = getManhattenHeuristic((nr, nc), goal_pos)

        tie_breaker += 1
        hq.heappush(pq, (new_hn, tie_breaker, (nr, nc), new_path, new_gn))

if final_path:
    print(f"Greedy Best-First Search Path found with {len(final_path)-1} steps.")
    print(f"Total Path Cost: {final_cost}")
    print("Coordinates:", final_path)
else:
    print("No valid path exists.")
```
