```python
import heapq as hq
import numpy as np
from collections import deque
```

# Question 1

```python
def readCube(filename):
    cube = []
    with open(filename, "r") as file:
        for line in file:
            row = line.strip().split()
            if row:
                cube.append(row)

    return np.array(cube)
```

# Question 2

```python
def bfs(cube):
    start = np.where(cube == "S")
    pos = (start[0][0], start[1][0])
    q = deque()

    q.append((pos, [pos]))

    sol = {pos}
    rows, cols = cube.shape

    while len(q) > 0:
        (r, c), path = q.popleft()

        if cube[r, c] == "G":
            return path

        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            if 0 <= nr < rows and 0 <= nc < cols:
                if cube[nr, nc] != "1" and (nr, nc) not in sol:
                    q.append(((nr, nc), path + [(nr, nc)]))
                    sol.add((nr, nc))

    return -1
```

```python
def dfs(cube):
    start = np.where(cube == "S")
    pos = (start[0][0], start[1][0])
    q = deque()

    q.append((pos, [pos]))

    sol = {pos}
    rows, cols = cube.shape

    while len(q) > 0:
        (r, c), path = q.pop()

        if cube[r, c] == "G":
            return path

        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            if 0 <= nr < rows and 0 <= nc < cols:
                if cube[nr, nc] != "1" and (nr, nc) not in sol:
                    q.append(((nr, nc), path + [(nr, nc)]))
                    sol.add((nr, nc))

    return -1
```

```python
def ucs(cube):
    start = np.where(cube == "S")
    pos = (start[0][0], start[1][0])
    q = []

    hq.heappush(q, (0, pos, [pos]))

    sol = {}
    rows, cols = cube.shape

    while len(q) > 0:
        cost, (r, c), path = hq.heappop(q)

        if cube[r, c] == "G":
            return cost, path

        if (r, c) in sol and sol[(r, c)] <= cost:
            continue
        sol[(r, c)] = cost

        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            if 0 <= nr < rows and 0 <= nc < cols:
                if cube[nr, nc] != "1" and (nr, nc) not in sol:
                    new_cost = cost + 1
                    hq.heappush(q, (new_cost, (nr, nc), path + [(nr, nc)]))

    return -1
```

```python
def ids(cube):
    for i in range(cube.shape[0] * cube.shape[1]):
        start = np.where(cube == "S")
        pos = (start[0][0], start[1][0])
        q = deque()

        limit = 0
        q.append((pos, [pos], limit))

        sol = {pos}
        rows, cols = cube.shape

        while len(q) > 0:
            (r, c), path, limit = q.pop()

            if limit > i:
                continue

            if cube[r, c] == "G":
                return path

            directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

            for dr, dc in directions:
                nr, nc = r + dr, c + dc

                if 0 <= nr < rows and 0 <= nc < cols:
                    if cube[nr, nc] != "1" and (nr, nc) not in sol:
                        nl = limit + 1
                        q.append(((nr, nc), path + [(nr, nc)], nl))
                        sol.add((nr, nc))

    return -1
```
