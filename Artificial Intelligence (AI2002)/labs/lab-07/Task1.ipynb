# Tic Tac Toe Game

```python
import copy
```

```python
EMPTY = None
```

```python
def initial_state():
    """
    Returns starting state of the board.
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]
```

```python
def draw_board(board):
    print("-------------")
    for i in range(3):
        print("| ", end="")
        for j in range(3):
            if board[i][j] == EMPTY:
                print(" ", end="")
            else:
                print(board[i][j], end="")
            print(" | ", end="")
        print()
        print("-------------")
```

```python
def player(board):
    """
    Returns player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return "O"
    return "X"
```

```python
def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res
```

```python
def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board
```

```python
def get_horizontal_winner(board):
    # check horizontally
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val
```

```python
def get_vertical_winner(board):
    # check vertically
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val
```

```python
def get_diagonal_winner(board):
    # check diagonally
    winner_val = None
    board_len = len(board)
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
    if winner_val:
        return winner_val

    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None

    return winner_val
```

```python
def winner(board):
    """
    Returns the winner of the game, if there is one.
    """
    winner_val = get_horizontal_winner(board) or get_vertical_winner(
        board) or get_diagonal_winner(board) or None
    return winner_val
```

```python
def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) is not None:
        return True

    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True
```

```python
def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == "X":
        return 1
    elif winner_val == "O":
        return -1
    return 0
```

```python
def alpha_beta_pruning(board):
    if terminal(board):
        return None

    current_player = player(board)
    best_move = None

    alpha = -float("inf")
    beta = float("inf")

    if current_player == "X":
        v = -float("inf")
        for action in actions(board):
            move_val = min_value(result(board, action), alpha, beta)
            if move_val > v:
                v = move_val
                best_move = action
            alpha = max(alpha, v)
    else:
        v = float("inf")
        for action in actions(board):
            move_val = max_value(result(board, action), alpha, beta)
            if move_val < v:
                v = move_val
                best_move = action
            beta = min(beta, v)

    return best_move
```

```python
def max_value(board, alpha, beta):
    if terminal(board):
        return utility(board)

    v = -float("inf")
    for action in actions(board):
        v = max(v, min_value(result(board, action), alpha, beta))
        alpha = max(alpha, v)
        if beta <= alpha:
            break

    return v
```

```python
def min_value(board, alpha, beta):
    if terminal(board):
        return utility(board)

    v = float("inf")
    for action in actions(board):
        v = min(v, max_value(result(board, action), alpha, beta))
        beta = min(beta, v)
        if beta <= alpha:
            break

    return v
```

```python
def play_game():
    board = initial_state()
    draw_board(board)

    while not terminal(board):
        valid_move = False
        while not valid_move:
            try:
                user_input = input(
                    "Enter row and col (0-2) separated by space: ").split()
                if len(user_input) != 2:
                    raise ValueError
                row, col = map(int, user_input)

                if (row, col) in actions(board):
                    valid_move = True
                else:
                    print("Invalid move! That spot is taken or out of bounds.")
            except ValueError:
                print("Please enter two numbers between 0 and 2.")

        board = result(board, (row, col))
        draw_board(board)

        if not terminal(board):
            print("AI is thinking...")
            action = alpha_beta_pruning(board)
            board = result(board, action)
            draw_board(board)

    res = winner(board)
    if res:
        print(f"Winner is {res}!")
    else:
        print("It's a draw.")
```

```python
if __name__ == "__main__":
    play_game()
```
