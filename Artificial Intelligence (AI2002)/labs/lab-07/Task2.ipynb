```python
import random
```

```python
def calculate_tour_cost(costs, tour):
    cost = 0
    for i in range(len(tour)):
        from_city = tour[i]

        to_city = tour[(i + 1) % len(tour)]
        cost += costs[from_city][to_city]

    return cost
```

```python
def get_neighbors(tour):
    neighbors = []
    for i in range(len(tour)):
        for j in range(i + 1, len(tour)):
            neighbor = tour.copy()
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighbors.append(neighbor)
    return neighbors
```

```python
def hill_climbing_tsp(costs):
    city_count = len(costs)

    current_tour = list(range(city_count))
    random.shuffle(current_tour)
    current_cost = calculate_tour_cost(costs, current_tour)

    while True:
        neighbors = get_neighbors(current_tour)

        best_neighbor = None
        best_neighbor_cost = float("inf")

        for neighbor in neighbors:
            cost = calculate_tour_cost(costs, neighbor)
            if cost < best_neighbor_cost:
                best_neighbor_cost = cost
                best_neighbor = neighbor

        if best_neighbor_cost < current_cost:
            current_cost = best_neighbor_cost
            current_tour = best_neighbor
        else:
            break

    return current_cost, current_tour
```

```python
if __name__ == "__main__":
    cost = [[0, 10, 15, 20],
            [10, 0, 35, 25],
            [15, 35, 0, 30],
            [20, 25, 30, 0]]

    min_distance, best_route = hill_climbing_tsp(cost)

    print(f"Best route found: {best_route}")
    print(f"Minimum total distance: {min_distance}")
```
